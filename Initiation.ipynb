#**4- Les filtres appliqués sur les images**

##**1- Gaussian Blurring**

```
blur = cv2.blur(img,(5,5))
plt.imshow(blur)
```

## **2- Median Blurring**

```
median = cv2.medianBlur(img,5)
plt.imshow(median)
```

## **3- Bilateral Blurring**

```
bilateral=cv2.bilateralFilter(img,9,75,75)
plt.imshow(bilateral)
```

#**5- Autres algorithmes appliqués sur les images**

## **1- Canny**

```
img = cv2.imread('messi5.jpg',0)
edges = cv2.Canny(img,100,200)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

## **2- Harris Corner Detection**

```
!wget -O mons.jpg https://upload.wikimedia.org/wikipedia/commons/a/ad/0_Mons_-_H%C3%B4tel_de_ville_%281%29.jpg
```

```
img = cv2.imread("mons.jpg")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)
dst = cv2.cornerHarris(gray,2,3,0.04)
#result is dilated for marking the corners, not important
dst = cv2.dilate(dst,None)
# Threshold for an optimal value, it may vary depending on the image.
img[dst>0.01*dst.max()]=[0,0,255]
plt.imshow(img)
```
